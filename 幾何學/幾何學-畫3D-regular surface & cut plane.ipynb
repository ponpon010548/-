```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
x = y = np.linspace(-4,4,50)
X, Y = np.meshgrid(x,y)
Z = 10*np.sin(X) + 5*Y**2
x_t, y_t = np.meshgrid(x,y)
z_t = 10*np.cos(X) + 10*Y - 10
fig = plt.figure()
ax = Axes3D(fig)
ax.plot_surface(x_t, y_t, z_t,color = 'red',alpha = '0.8')
ax.plot_surface(X,Y,Z,alpha = '0.7')
```
