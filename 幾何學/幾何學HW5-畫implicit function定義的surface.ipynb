```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
K = np.linspace(-9.9, 9.9 , 100)
x = y = np.linspace(-10, 10 ,100)
X, Y = np.meshgrid(x, y)
```

```python
ax = plt.gca(projection = '3d')
ax.set_xlim(-10,10)
ax.set_ylim(-10,10)
ax.set_zlim(-10,10)

for z in K:
    Z = np.exp(X) + 3 * np.cos(Y) + z**3
    plt.contour(X, Y, Z+z, [z] ,zdir = 'z')
```
